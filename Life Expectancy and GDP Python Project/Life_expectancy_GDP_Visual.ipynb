```python
#import 
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#download data and looked at it
df = pd.read_csv("all_data.csv")
df.head()
```

```python
#found the dimensions
df.shape
```

```python
#found the details
print(df.Country.unique())
print(df.Year.min())
print(df.Year.max())
print(df.Year.unique())
```

```python
#renamed column to make things easier
df = df.rename({"Life expectancy at birth (years)":"LEABY"}, axis = "columns")
df.head()
```

```python
#renamed United States of America to USA because it wasn't fitting on the labels/titles
df.Country.replace('United States of America', 'USA', inplace=True)
```

```python
#created a dataframe of the averages by country
dfMeans = df.drop("Year", axis = 1).groupby("Country").mean().reset_index()
dfMeans.head(6)
```

```python
#GDP Bar Graph
plt.figure(figsize=(8,6))
plt.bar(dfMeans.Country, dfMeans.GDP)
plt.xlabel("GDP in Trillions of U.S. Dollars")
plt.title('GDP Bar Graph')
```

```python
# Life Expectancy Bar Graph
plt.figure(figsize=(8,6))
plt.bar(dfMeans.Country, dfMeans.LEABY)
plt.xlabel("Life Expectancy by Country")
plt.title('Life Expectancy Bar Graph')
```

```python
# created a dataframe of only Chile so I could plot it easier
dfchile = df[df.Country == 'Chile']
dfchile.shape
```

```python
#created dataframes for the remaining nations
dfchina = df[df.Country == 'China']
dfgermany = df[df.Country == 'Germany']
dfmexico = df[df.Country == 'Mexico']
dfusa = df[df.Country == 'USA']
dfzimbabwe = df[df.Country == 'Zimbabwe']
```

```python
#GDP Line Graph
plt.figure(figsize=(8,6))
plt.plot(dfchile.Year,dfchile.GDP, color ='Orange', linestyle = '-', label = "Chile")
plt.plot(dfchina.Year,dfchina.GDP, color ='Red', linestyle = '-', label = "China")
plt.plot(dfgermany.Year,dfgermany.GDP, color ='Black', linestyle = '-', label = "Germany")
plt.plot(dfmexico.Year,dfmexico.GDP, color ='Green', linestyle = '-', label = "Mexico")
plt.plot(dfusa.Year,dfusa.GDP, color ='Blue', linestyle = '-', label = "USA")
plt.plot(dfzimbabwe.Year,dfzimbabwe.GDP, color ='Purple', linestyle = '-', label = "Zimbabwe")
plt.legend()
plt.xlabel('Year')
plt.ylabel("GDP in Trillions of U.S. Dollars")
plt.title('GDP Line Graph')
plt.show()
```

```python
#GDP Line Graphs
plt.subplot(2, 3, 1)
plt.plot(dfchile.Year,dfchile.GDP, color ='Orange', linestyle = '-')
plt.title("Chile")
plt.subplot(2, 3, 2)
plt.plot(dfchina.Year,dfchina.GDP, color ='Red', linestyle = '-')
plt.title("China")
plt.subplot(2, 3, 3)
plt.plot(dfgermany.Year,dfgermany.GDP, color ='Black', linestyle = '-')
plt.title("Germany")
plt.subplot(2, 3, 4)
plt.plot(dfmexico.Year,dfmexico.GDP, color ='Green', linestyle = '-')
plt.title("Mexico")
plt.subplot(2, 3, 5)
plt.plot(dfusa.Year,dfusa.GDP, color ='Blue', linestyle = '-')
plt.title("USA")
plt.subplot(2, 3, 6)
plt.plot(dfzimbabwe.Year,dfzimbabwe.GDP, color ='Purple', linestyle = '-')
plt.title("Zimbabwe")
plt.tight_layout()
plt.suptitle("GDP Line Graphs")
plt.show()
```

```python
#Life Expectancy Line Graph
plt.figure(figsize=(8,6))
plt.plot(dfchile.Year,dfchile.LEABY, color ='Orange', linestyle = '-', label = "Chile")
plt.plot(dfchina.Year,dfchina.LEABY, color ='Red', linestyle = '-', label = "China")
plt.plot(dfgermany.Year,dfgermany.LEABY, color ='Black', linestyle = '-', label = "Germany")
plt.plot(dfmexico.Year,dfmexico.LEABY, color ='Green', linestyle = '-', label = "Mexico")
plt.plot(dfusa.Year,dfusa.LEABY, color ='Blue', linestyle = '-', label = "USA")
plt.plot(dfzimbabwe.Year,dfzimbabwe.LEABY, color ='Purple', linestyle = '-', label = "Zimbabwe")
plt.legend()
plt.xlabel('Year')
plt.ylabel("Life Expectancy")
plt.title('Life Expectancy Line Graph')
plt.show()
```

```python
#Life Expectancy Line Graphs
plt.subplot(2, 3, 1)
plt.plot(dfchile.Year,dfchile.LEABY, color ='Orange', linestyle = '-')
plt.title("Chile")
plt.subplot(2, 3, 2)
plt.plot(dfchina.Year,dfchina.LEABY, color ='Red', linestyle = '-')
plt.title("China")
plt.subplot(2, 3, 3)
plt.plot(dfgermany.Year,dfgermany.LEABY, color ='Black', linestyle = '-')
plt.title("Germany")
plt.subplot(2, 3, 4)
plt.plot(dfmexico.Year,dfmexico.LEABY, color ='Green', linestyle = '-')
plt.title("Mexico")
plt.subplot(2, 3, 5)
plt.plot(dfusa.Year,dfusa.LEABY, color ='Blue', linestyle = '-')
plt.title("USA")
plt.subplot(2, 3, 6)
plt.plot(dfzimbabwe.Year,dfzimbabwe.LEABY, color ='Purple', linestyle = '-')
plt.title("Zimbabwe")
plt.tight_layout()
plt.suptitle("Life Expectancy Line Graphs")
plt.show()
```
